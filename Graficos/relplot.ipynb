```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
def grafico_relplot(data= object, 
                    x= str, y= str, 
                    hue_style= str, 
                    g_titulo= str, 
                    x_titulo= str, 
                    y_titulo= str, 
                    l_titulo= str, 
                    palette='rocket',
) -> object:
    
    grafico = sns.relplot(data=data, x=x, y=y, hue=hue_style, palette=palette,)
    grafico.ax.set_title(label=g_titulo, fontsize=12, fontweight='bold')
    grafico.set_xlabels(label=x_titulo, fontsize=10)
    grafico.set_ylabels(label=y_titulo, fontsize=10)
    grafico.legend.set_title(title=l_titulo, prop={'size': 10, 'weight': 'bold'})
    grafico.figure.set_size_inches(w=10, h=6)

    return grafico
```

```python
grafico1 = grafico_relplot(
        data=data, 
        x='total_bill', 
        y='tip', 
        hue_style='day', 
        g_titulo='Gorjetas', 
        x_titulo='Valor da conta (USD)', 
        y_titulo='Valor da gorjeta (USD)', 
        l_titulo='Dias da semana'
)
```

```python
grafico1.figure.savefig(fname='relplot_gorjetas.png', bbox_inches='tight')
```

```python
data_diamonds = sns.load_dataset('diamonds')
data_diamonds.head()
```

```python
grafico2 = grafico_relplot(
    data=data_diamonds,
    x='price',
    y='carat',
    hue_style='cut',
    g_titulo='Diamantes',
    x_titulo='Preço',
    y_titulo='Quilate',
    l_titulo='Corte',
)
```

```python
grafico2.figure.savefig(fname='relplot_diamantes.png', bbox_inches='tight')
```
